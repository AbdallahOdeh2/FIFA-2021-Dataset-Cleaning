## **create function to convert height**

```python
def convert_height_cm(x):
    if 'cm' in x:
        return int(x.replace('cm', ''))
    else:
        height = x.split("'")
        return round(int(height[0])*30.48+int(height[1][0])*2.54)
```

## **create function to convert weight**

```python
def weight_to_kg(x):
    if "kg" in x:
        return int(x[:-2])
    else:
        return round(int(x[:-3])*0.45359237)
```

## **apply the functions the have been created**

```python
df['Weight'] = df['Weight'].apply(weight_to_kg)
df['Height'] = df['Height'].apply(convert_height_cm)
```

## **remove the extra spaces**

```python
df['Club'] = df['Club'].str.strip()
```

## **convert joined to date then extract the year,month,day**

```python
df['Joined'] = pd.to_datetime(df['Joined'])
df['Year_Joined'] = df['Joined'].dt.year
df['Month_Joined'] = df['Joined'].dt.month
df['Day_Joined'] = df['Joined'].dt.day
```

## **create function to convert some columns to numeric**

```python
def convert_value(x):
  x = x.replace('€',"")
  if 'M' in x:
    return int(float(x[:-1])*1000*1000)
  elif 'K' in x:
    return int(float(x[:-1])*1000)
  else:
    return round(float(x))
```

## **apply the function on some [Value,Wage,Release Clause]**

```python
df['Value'] = df['Value'].apply(convert_value)
df['Wage'] = df['Wage'].apply(convert_value)
df['Release Clause'] = df['Release Clause'].apply(convert_value)
```

## **convert [IR,SM,W/F] to numeric**

```python
df['IR'] = df['IR'].str.strip('★').astype(int)
df['SM'] = df['SM'].str.strip('★').astype(int)
df['W/F'] = df['W/F'].str.strip('★').astype(int)
```

## **Count the duration since the player has been joined**

```python
df['Membership_Duration'] = 2023-df.Year_Joined
```
